```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Iris Dataset.csv")

# Features and labels
X = df.iloc[:, :-1]  # assuming last column is the label
y = df.iloc[:, -1]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Try different values of K
k_values = [1, 3, 5, 7]
models = {}
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    models[k] = knn
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

for k, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    print(f"\nK={k} | Accuracy: {acc:.2f}")
    print("Confusion Matrix:\n", cm)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification # Added for sample data
from sklearn.preprocessing import StandardScaler # Added for scaling

# --- Start of added code to create a sample dataset ---
# Generate a sample dataset for demonstration
X, y = make_classification(
    n_samples=200,          # 200 data points
    n_features=2,           # 2 features for easy visualization
    n_informative=2,
    n_redundant=0,
    n_classes=3,            # 3 classes
    n_clusters_per_class=1,
    random_state=42
)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# --- End of added code ---

# Reduce to 2 features for visualization
X_vis = X_scaled[:, :2]
X_train_vis, X_test_vis, y_train_vis, y_test_vis = train_test_split(
    X_vis, y, test_size=0.3, random_state=42
)

def plot_decision_boundary(knn, X, y, k):
    h = 0.02  # step size in the mesh
    cmap_light = ListedColormap(['#FFC0CB', '#90EE90', '#ADD8E6']) # A slightly different colormap
    cmap_bold = ['#8B0000', '#006400', '#00008B'] # Darker, more distinct colors

    # Fit classifier
    knn.fit(X, y)

    # Mesh grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )

    # Predict over mesh
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.6)

    # Plot training points
    for idx, class_color in enumerate(cmap_bold):
        plt.scatter(
            X[y == idx, 0], X[y == idx, 1],
            c=class_color, label=f"Class {idx}",
            edgecolor='k', s=70
        )

    plt.title(f"KNN Decision Boundary (K={k})", fontsize=16)
    plt.xlabel("Feature 1", fontsize=12)
    plt.ylabel("Feature 2", fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Visualize for K=3
plot_decision_boundary(KNeighborsClassifier(n_neighbors=3), X_train_vis, y_train_vis, k=3)
```
